```python
final_df=df_pur_pl.pivot_table(index='game_name',columns='user_id',values='hours').fillna(0)
final_df.head()
```

```python
final_df_matrix = csr_matrix(final_df.values)
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(final_df_matrix)
```

```python
final_df.shape
```

```python
query_index = 2210
print(query_index)
distances, indices = model_knn.kneighbors(final_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(final_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}'.format(i, final_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
